```python
import pandas as pd
```

```python
spot = pd.read_csv("../data/nifty_spot_clean.csv")
futures = pd.read_csv("../data/nifty_futures_clean.csv")
options = pd.read_csv("../data/nifty_options_clean.csv")
```

```python
for df in [spot, futures, options]:
    df["datetime"] = pd.to_datetime(df["datetime"])
```

```python
spot = spot.rename(columns={
    "open": "spot_open",
    "high": "spot_high",
    "low": "spot_low",
    "close": "spot_close",
    "volume": "spot_volume"
})

futures = futures.rename(columns={
    "open": "fut_open",
    "high": "fut_high",
    "low": "fut_low",
    "close": "fut_close",
    "volume": "fut_volume"
})

options = options.rename(columns={
    "open": "opt_open",
    "high": "opt_high",
    "low": "opt_low",
    "close": "opt_close",
    "volume": "opt_volume"
})


```

```python
merged = pd.merge(
    spot,
    futures,
    on="datetime",
    how="inner"
)

merged = pd.merge(
    merged,
    options,
    on="datetime",
    how="inner"
)

```

```python
merged.to_csv("../data/nifty_merged_5min.csv", index=False)
```
