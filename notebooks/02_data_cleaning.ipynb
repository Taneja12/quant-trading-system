```python
import pandas as pd
import numpy as np

spot = pd.read_csv("../data/nifty_spot_5min.csv")
futures = pd.read_csv("../data/nifty_futures_5min.csv")
options = pd.read_csv("../data/nifty_options_5min.csv")
```

```python
for df in [spot, futures, options]:
    df["datetime"] = pd.to_datetime(df["datetime"])
```

```python
spot = spot.sort_values("datetime").drop_duplicates("datetime")
futures = futures.sort_values("datetime").drop_duplicates("datetime")
options = options.sort_values("datetime").drop_duplicates("datetime")
```

```python
spot = spot.ffill()
futures = futures.ffill()
options = options.ffill()
```

```python
def filter_market_hours(df):
    return df[
        (df["datetime"].dt.time >= pd.to_datetime("09:15").time()) &
        (df["datetime"].dt.time <= pd.to_datetime("15:30").time())
    ]

spot = filter_market_hours(spot)
futures = filter_market_hours(futures)
options = filter_market_hours(options)
```

```python
spot.to_csv("../data/nifty_spot_clean.csv", index=False)
futures.to_csv("../data/nifty_futures_clean.csv", index=False)
options.to_csv("../data/nifty_options_clean.csv", index=False)
```
