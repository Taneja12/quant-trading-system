```python
import pandas as pd

df = pd.read_csv("../data/nifty_spot_raw.csv")
df.head()
df.info()
```

```python
# convert date column to datetime
df["datetime"] = pd.to_datetime(df["date"])

# sort data
df = df.sort_values("datetime")

# set datetime as index
df = df.set_index("datetime")

df.head()

```

```python
df.head()
df.info()
```

```python
df_5min = df.resample("5min").agg({
    "open": "first",
    "high": "max",
    "low": "min",
    "close": "last",
    "volume": "sum"
})
```

```python
df_5min = df_5min.dropna()
df_5min = df_5min.reset_index()

df_5min.info()
```

```python
df_5min = df_5min[
    df_5min["datetime"] >= df_5min["datetime"].max() - pd.DateOffset(years=1)
]

df_5min.info()
```

```python
# Spot data
df_5min.to_csv("../data/nifty_spot_5min.csv", index=False)

# Futures & Options (approximated)
df_5min.to_csv("../data/nifty_futures_5min.csv", index=False)
df_5min.to_csv("../data/nifty_options_5min.csv", index=False)
```
